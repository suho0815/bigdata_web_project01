```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('C:/새 폴더/Uploads/pet_찐찐최종.csv', encoding='cp949')

def split_address(address):
    address_parts = address.split()
    si = address_parts[0]
    rest = address_parts[1:]
    
    idx = next((i for i, x in enumerate(rest) if re.search('[시군구]$', x)), None)
    
    if idx is not None:
        gun_gu = ' '.join(rest[:idx+1])
        eup_myeon_dong = ' '.join(rest[idx+1:])
    else:
        gun_gu = None
        eup_myeon_dong = ' '.join(rest)
    
    return pd.Series([si, gun_gu, eup_myeon_dong])

data['address'] = data['address'].fillna("Unknown")

data[['Si', 'Gun_Gu', 'Eup_Myeon_Dong']] = data['address'].apply(split_address)

data.head()
```

```python
data = data.drop(['address'], axis=1)
```

```python
data.tail(20)
```

```python
data = data.dropna(subset = ['longitude', 'latitude'])
```

```python
data.rename(columns= {'Si':'Si/Do', 'Gun_Gu':'Si/Gun/Gu', 'Eup_Myeon_Dong':'Eup/Myeon/Dong'}, inplace=True)
```

```python
data = data.replace('Unknown', pd.NaT) 
data = data.replace('?', pd.NaT)
data = data.dropna()
```
